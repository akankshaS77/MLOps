```python
df.groupby(['Card Type'])['Is Fraudulent'].mean()
Mean_encoded = df.groupby(['Card Type'])['Is Fraudulent'].mean().to_dict()
df['Card Type'] = df['Card Type'].map(Mean_encoded)
Mean_encoded = df.groupby(['MCC Category'])['Is Fraudulent'].mean().to_dict()
df['MCC Category'] = df['MCC Category'].map(Mean_encoded)
Mean_encoded = df.groupby(['Device'])['Is Fraudulent'].mean().to_dict()
df['Device'] = df['Device'].map(Mean_encoded)
Mean_encoded = df.groupby(['Location'])['Is Fraudulent'].mean().to_dict()
df['Location'] = df['Location'].map(Mean_encoded)
Mean_encoded = df.groupby(['Merchant Reputation'])['Is Fraudulent'].mean().to_dict()
df['Merchant Reputation'] = df['Merchant Reputation'].map(Mean_encoded)
Mean_encoded = df.groupby(['Online Transactions Frequency'])['Is Fraudulent'].mean().to_dict()
df['Online Transactions Frequency'] = df['Online Transactions Frequency'].map(Mean_encoded)
df.drop('Date', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
# create a MinMaxScaler object
scaler = MinMaxScaler()

# fit and transform the data
normalized_data = scaler.fit_transform(df.drop('Is Fraudulent', axis=1))
```

```python
from pandas import DataFrame
# convert the array back to a dataframe
df_norm = DataFrame(normalized_data)
df_norm.head()
```

```python
import pandas as pd
import numpy as np
from faker import Faker

# Initialize Faker for generating realistic text data
fake = Faker()

# Number of additional rows to generate
additional_rows = 5_000

# Parameters for balancing
fraud_ratio = 0.5  # Desired ratio for fraudulent cases
fraudulent_rows = int(additional_rows * fraud_ratio)
non_fraudulent_rows = additional_rows - fraudulent_rows

# Generate synthetic data
def generate_transaction_data(n_rows, is_fraudulent):
    data = {
        "Unnamed: 0": np.arange(1, n_rows + 1),  # Unique IDs
        "Amount": np.random.uniform(10, 5000, size=n_rows).round(2),  # Random transaction amounts
        "Previous Transactions": np.random.randint(0, 100, size=n_rows),  # Number of previous transactions
        "Balance Before Transaction": np.random.uniform(1000, 50000, size=n_rows).round(2),  # Account balance
        "Time of Day": np.random.choice(["Morning", "Afternoon", "Evening", "Night"], size=n_rows),  # Random time
        "Velocity": np.random.uniform(0, 1, size=n_rows).round(2),  # Velocity score
        "Customer Age": np.random.randint(18, 70, size=n_rows),  # Random ages
        "Customer Income": np.random.uniform(20000, 200000, size=n_rows).round(2),  # Customer income
        "Card Limit": np.random.uniform(1000, 50000, size=n_rows).round(2),  # Credit card limit
        "Credit Score": np.random.randint(300, 850, size=n_rows),  # Credit score
        "Merchant Location History": [fake.city() for _ in range(n_rows)],  # Merchant city
        "Spending Patterns": np.random.choice(["Conservative", "Moderate", "Aggressive"], size=n_rows),  # Patterns
        "Is Fraudulent": [is_fraudulent] * n_rows,  # Fraudulent flag
    }
    return pd.DataFrame(data)

# Generate fraudulent and non-fraudulent datasets
fraudulent_data = generate_transaction_data(fraudulent_rows, 1)  # Fraudulent transactions
non_fraudulent_data = generate_transaction_data(non_fraudulent_rows, 0)  # Non-fraudulent transactions

# Combine the two datasets
balanced_data = pd.concat([fraudulent_data, non_fraudulent_data]).reset_index(drop=True)

# Shuffle the data
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Save to a CSV file
output_file = "C:\\Users\\Aorus\\OneDrive\\Desktop\\MLOps\\MLOps_2\\MLOps\\data\\balanced_dataset.csv"
balanced_data.to_csv(output_file, index=False)

print(f"Balanced dataset with {additional_rows} rows saved to '{output_file}'")
```
